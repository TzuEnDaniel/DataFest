```python
import time
import json
import requests
import pandas as pd
from pyecharts import options as opts
from pyecharts.charts import Bar, Timeline, Grid
from pyecharts.globals import ThemeType
```

```python
import pandas as pd
data=pd.read_csv('C:\\Users\\CloudMosa\\Desktop\\Penn state\\DataFest\\data\\GroupYearDesc.csv',encoding='gbk')
data.head(10)
```

```python
tl = Timeline({"theme": ThemeType.MACARONS})
for i in range(11):
    bar = (
        Bar({"theme": ThemeType.MACARONS})
        .add_xaxis(list(data['Category'])[i*10:i*10+10][::-1])
        .add_yaxis("Law Event", list(data['count'])[i*10:i*10+10][::-1])
        .reversal_axis()
        .set_global_opts(
            title_opts=opts.TitleOpts("{}".format(list(data['Year'])[i*10]),pos_right='0%',pos_bottom='15%'),
            xaxis_opts=opts.AxisOpts(
                splitline_opts=opts.SplitLineOpts(is_show=True)),
            yaxis_opts=opts.AxisOpts(splitline_opts=opts.SplitLineOpts(is_show=True),
                                     axislabel_opts=opts.LabelOpts(color='#FF7F50')),)
        .set_series_opts(label_opts=opts.LabelOpts(position="right",color='#9400D3'))
    )
    grid = (
        Grid()
        .add(bar, grid_opts=opts.GridOpts(pos_left="20%",pos_right="0%"))
    )
    tl.add(grid, "{}".format(data['Year'][i*10]))  #设置标签
    tl.add_schema(
        play_interval=300,   #播放速度
        is_timeline_show=True,  #是否显示 timeline 组件
        is_auto_play=False,
    )
tl.render_notebook()
```
